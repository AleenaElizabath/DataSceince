```python
import numpy as np
import matplotlib.pyplot as mpt
ar1=np.array([1,5,9,4,11,6])
ar3=np.array([11,5,19,54,11,6])
ar2=np.array([81,24,33,44,25,46])
ar4=np.array([8,24,3,44,65,46])
#triangle as a marker 
mpt.plot(ar1,marker='^')
mpt.plot(ar2,marker='o')
#labeling x and y axis
mpt.xlabel("distance")
mpt.ylabel("time")
#giving title of the graph and adding gridlines
mpt.title("Distance-time graph")
#axis="x" for grid only for x axis and axis="y" for grid inly for y axis
mpt.grid(axis="y",color='g')
#adding clr for the graph
#giving subplots(dividing the graph into 2 or 4 or...so on)
```

```python
mpt.suptitle("Plotting 4 subgraphs")
ar1=np.array([1,5,9,4,11,6])
ar3=np.array([11,5,19,54,11,6])
ar2=np.array([81,24,33,44,25,46])
ar4=np.array([8,24,3,44,65,46])
mpt.subplot(2,2,1)
mpt.plot(ar1,color='red',marker='o')
mpt.xlabel("distance")
mpt.ylabel("time")
mpt.subplot(2,2,2)
mpt.plot(ar2,marker='d')
mpt.xlabel("velocity")
mpt.ylabel("time")
mpt.subplot(2,2,3)
mpt.plot(ar3,color='green',marker='^')
mpt.xlabel("speed")
mpt.ylabel("time")
mpt.subplot(2,2,4)
mpt.plot(ar4,color='violet',marker='s')
mpt.xlabel("acceleration")
mpt.ylabel("time")
```

```python
#scatterplot
mpt.suptitle("age vs BMI")
age=np.array([1,5,9,4,11,6])
bmi=np.array([11,5,19,54,11,6])
mpt.scatter(age,bmi)
mpt.scatter(ar1,ar2)
mpt.scatter(ar3,ar4)
```

```python
#labelling and giving diff clr to diff outcome using seaborn
import seaborn as sb
age=np.array([1,5,9,4,11,6])
bmi=np.array([11,5,19,54,11,5])
sugar=np.array(['yes','no','yes','no','yes','no'])
sb.scatterplot(x=bmi,y=age,hue=sugar,palette={'yes':'red','no':'blue'})
```
